```python
# Question 2 - Extracting Tesla Revenue Data Using Webscraping
import requests
from bs4 import BeautifulSoup
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
tsla_url = "https://stockanalysis.com/stocks/tsla/revenue/"
tsla_revenue_data = requests.get(tsla_url).text
soup = BeautifulSoup(tsla_data, 'html.parser')
```

```python
tsla_revenue_data = pd.DataFrame(columns=["Fiscal year end","Revenue"])
for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    Fiscal_year_end = col[0].text
    Revenue = col[1].text
    tsla_revenue_data = pd.concat([tsla_data,pd.DataFrame({"Fiscal year end":[Fiscal_year_end], "Revenue":[Revenue]})], ignore_index=True)
tsla_revenue_data.head()
```

```python
# Question 3 - Extracting GameStop Stock Data Using yfinance
GME = yf.Ticker("GME")
GME_stock = GME.history(period='max')
GME_stock
GME_stock.reset_index(inplace=True)
GME_stock.head()
```

```python
# Question 4 - Extracting GameStop Revenue Data Using Webscraping 
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
GME_url = "https://stockanalysis.com/stocks/gme/revenue/"
GME_revenue_data = requests.get(GME_url).text
soup_gme = BeautifulSoup(GME_revenue_data, 'html.parser')
GME_revenue_data = pd.DataFrame(columns=["Fiscal year end","Revenue"])
for row in soup_gme.find("tbody").find_all("tr"):
    col = row.find_all("td")
    Fiscal_year_end = col[0].text
    Revenue = col[1].text
    GME_revenue_data = pd.concat([GME_revenue_data,pd.DataFrame({"Fiscal year end":[Fiscal_year_end], "Revenue":[Revenue]})], ignore_index=True)
GME_revenue_data.head()
```

```python
# Question 5 - Tesla Stock and Revenue Dashboard
import matplotlib.pyplot as plt
def make_graph(stock_data, revenue_data, title="Tesla Stock vs Revenue", x_label="stock", y_label="Revenue"):
    if len(stock_data) != len(revenue_data):
        raise ValueError("The length of stock_data and revenue_data must be the same.")
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data, revenue_data, marker="o", label=f"{y_label} over {x_label}")
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.show()
make_graph(tsla_stock["Close"].head(), tsla_revenue_data['Revenue'].head())
```
